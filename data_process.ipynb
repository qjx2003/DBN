```python
# 读取 Excel 文件
df = pd.read_excel('training_dropped.xlsx')

# 清理列名空格
df.columns = df.columns.str.strip()

# 转换 Completion Date 为 datetime 格式
df['Completion Date'] = pd.to_datetime(df['Completion Date'], errors='coerce')

# 删除 Completion Date 为缺失的行
df = df.dropna(subset=['Completion Date'])

# 按 Match ID 和 Completion Date 排序
df = df.sort_values(['Match ID 18Char', 'Completion Date']).reset_index(drop=True)

# 分组计算时间差
df['Days Since Last Completion'] = (
    df.groupby('Match ID 18Char')['Completion Date']
    .diff()
    .dt.days
    .fillna(0)
    .astype(int)
)
```
